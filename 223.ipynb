```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.mixture import GaussianMixture
import random

rand_state = random.randint(00000, 99999)
np.random.seed(rand_state)

def viz(data, mu, sd):
    GR = '#40909A'
    OR = '#C76C2B'
    ALP = 0.5
    plt.figure()

    data_min = min(min(data[0]), min(data[1]))
    data_max = max(max(data[0]), max(data[1]))

    x = np.linspace(data_min, data_max, 100)
    p = norm.pdf(x, mu[0], sd[0])
    plt.plot(x, p, linewidth=5, color=GR)

    x = np.linspace(data_min, data_max, 100)
    p = norm.pdf(x, mu[1], sd[1])
    plt.plot(x, p, linewidth=5, color=OR)

    _ = plt.hist(data[0], bins=10, density=True, color=GR, alpha=ALP)
    _ = plt.hist(data[1], bins=10, density=True, color=OR, alpha=ALP)

    plt.plot(mu[0], 0, linestyle='None', marker='o', markersize=10, alpha=ALP, color=GR)
    plt.plot(mu[1], 0, linestyle='None', marker='o', markersize=10, alpha=ALP, color=OR)

    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.show()

mu = [1, 3]
sd = [0.5, 0.35]
data1 = np.random.normal(mu[0], sd[0], size=100)
data2 = np.random.normal(mu[1], sd[1], size=100)
viz([list(data1)]+[list(data2)], mu, sd)

data = np.expand_dims(np.concatenate([data1, data2]), axis=1)
gmm = GaussianMixture(n_components=2, init_params='random')
gmm.fit(data)

print('GMM >>>')
w1, w2 = gmm.weights_  # Q1
mu1, mu2 = gmm.means_.flatten()  # Q2
sd1, sd2 = np.sqrt(gmm.covariances_).flatten()  # Q3
print('p(x) = %.2f * Normal(%.2f, %.2f) + %.2f * Normal(%.2f, %.2f)' % (w1, mu1, sd1, w2, mu2, sd2))
viz([list(data1)] + [list(data2)], [mu1, mu2], [sd1, sd2])
```
